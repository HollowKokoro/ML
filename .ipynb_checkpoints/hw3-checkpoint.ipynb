players.csv

```python
def switch_case_foot(arg, res1, res2):
    if arg == 'right':
        res1 = 1
    elif arg == 'left':
        res2 = 1
    return res1, res2

def switch_case_intensity(arg, res1, res2, res3):
    if arg == 'low':
        res1 = 1
    elif arg == 'medium':
        res2 = 1
    elif arg == 'high':
        res3 = 1
    return res1, res2, res3
```

```python
df_players = pd.read_csv('players.csv')
df_players = df_players.drop_duplicates()
df_players['preferred_right_foot'] = 0
df_players['preferred_left_foot'] = 0
df_players['low_attacking_work_rate'] = 0
df_players['medium_attacking_work_rate'] = 0
df_players['high_attacking_work_rate'] = 0
df_players['low_defensive_work_rate'] = 0
df_players['medium_defensive_work_rate'] = 0
df_players['high_defensive_work_rate'] = 0

for i in df_players.index:
    df_players['preferred_right_foot'].iloc[i], df_players['preferred_left_foot'].iloc[i] =\
        switch_case_foot(
            df_players['preferred_foot'].iloc[i],
            df_players['preferred_right_foot'].iloc[i],
            df_players['preferred_left_foot'].iloc[i]
        )
    df_players['low_attacking_work_rate'].iloc[i], df_players['medium_attacking_work_rate'].iloc[i], df_players['high_attacking_work_rate'].iloc[i] =\
        switch_case_intensity(
            df_players['attacking_work_rate'].iloc[i],
            df_players['low_attacking_work_rate'].iloc[i],
            df_players['medium_attacking_work_rate'].iloc[i],
            df_players['high_attacking_work_rate'].iloc[i]
        )
    df_players['low_defensive_work_rate'].iloc[i], df_players['medium_defensive_work_rate'].iloc[i], df_players['high_defensive_work_rate'].iloc[i] =\
        switch_case_intensity(
            df_players['defensive_work_rate'].iloc[i],
            df_players['low_defensive_work_rate'].iloc[i],
            df_players['medium_defensive_work_rate'].iloc[i],
            df_players['high_defensive_work_rate'].iloc[i]
        )
df_players.head()
```
