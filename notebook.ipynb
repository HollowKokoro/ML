Gensim - это библиотека Python для моделирования тем, индексации документов и поиска сходства с большими корпусами. В ней до 4 версии была функция суммаризации, но в 4 версии удалена. Здесь используется последняя версия до 4

```python
short_summary = summarize(original_text)
print(short_summary)
```

```python
parser = PlaintextParser.from_string(original_text,Tokenizer('english'))
```

LexRank использует графический подход на основе центральности собственных векторов. Предложения помещаются в вершины графов, веса на краях рассчитываются с использованием метрики косинусного сходства.

```python
lex_rank_summarizer = LexRankSummarizer()
lexrank_summary = lex_rank_summarizer(parser.document,sentences_count=3)
lex = ""

for sentence in lexrank_summary:
    lex += " " + str(sentence)
print(lex)

lex_score = Rouge().get_scores(original_text, lex)
rouge_1_lex_score = lex_score[0]['rouge-1']['p']
rouge_2_lex_score = lex_score[0]['rouge-2']['p']
rouge_l_lex_score = lex_score[0]['rouge-l']['p']
```

Скрытый семантический анализ (LSA) обычно используется для обнаружения темы. Он основан на SVD (разложение по сингулярным значениям).

```python
lsa_summarizer=LsaSummarizer()
lsa_summary= lsa_summarizer(parser.document,sentences_count=3)
lsa = ""

for sentence in lsa_summary:
    lsa += " " + str(sentence)
print(lsa)

lsa_score = Rouge().get_scores(original_text, lsa)
rouge_1_lsa_score = lsa_score[0]['rouge-1']['p']
rouge_2_lsa_score = lsa_score[0]['rouge-2']['p']
rouge_l_lsa_score = lsa_score[0]['rouge-l']['p']
```

Алгоритм Луна оценивает «значимость» слова, вычисляемую по частоте.

```python
luhn_summarizer=LuhnSummarizer()
luhn_summary=luhn_summarizer(parser.document,sentences_count=3)
luhn = ""

for sentence in luhn_summary:
    luhn += " " + str(sentence)
print(luhn)

luhn_score = Rouge().get_scores(original_text, luhn)
rouge_1_luhn_score = luhn_score[0]['rouge-1']['p']
rouge_2_luhn_score = luhn_score[0]['rouge-2']['p']
rouge_l_luhn_score = luhn_score[0]['rouge-l']['p']
```

Алгоритм суммирования Кульбака-Либера (KL) фокусируется на минимизации суммарного словаря путем проверки отклонения от входного словаря.

```python
kl_summarizer=KLSummarizer()
kl_summary=kl_summarizer(parser.document,sentences_count=3)
kl = ""

for sentence in kl_summary:
    kl += " " + str(sentence)
print(kl)

kl_score = Rouge().get_scores(original_text, kl)
rouge_1_kl_score = kl_score[0]['rouge-1']['p']
rouge_2_kl_score = kl_score[0]['rouge-2']['p']
rouge_l_kl_score = kl_score[0]['rouge-l']['p']
```

```python
# Loading the model and tokenizer for bart-large-cnn
tokenizer=BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model=BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

# Encoding the inputs and passing them to model.generate()
inputs = tokenizer.batch_encode_plus([original_text],return_tensors='pt')
summary_ids = model.generate(inputs['input_ids'], early_stopping=True)

# Decoding and printing the summary
bart_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(bart_summary)

bart_score = Rouge().get_scores(original_text, bart_summary)
rouge_1_bart_score = bart_score[0]['rouge-1']['p']
rouge_2_bart_score = bart_score[0]['rouge-2']['p']
rouge_l_bart_score = bart_score[0]['rouge-l']['p']
```

```python
# Instantiating the model and tokenizer with gpt-2
tokenizer=GPT2Tokenizer.from_pretrained('gpt2')
model=GPT2LMHeadModel.from_pretrained('gpt2')

# Encoding text to get input ids & pass them to model.generate()
inputs = tokenizer.batch_encode_plus([original_text], return_tensors='pt', max_length=512)
summary_ids=model.generate(inputs['input_ids'], early_stopping=True)

# Decoding and printing summary
GPT_summary=tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(GPT_summary)

GPT_score = Rouge().get_scores(original_text, GPT_summary)
rouge_1_GPT_score = bart_score[0]['rouge-1']['p']
rouge_2_GPT_score = bart_score[0]['rouge-2']['p']
rouge_l_GPT_score = bart_score[0]['rouge-l']['p']
```

```python
df = DataFrame({
    "lex" : [rouge_1_lex_score],
    "lsa" : [rouge_1_lsa_score],
    "luhn" : [rouge_1_luhn_score],
    "kl" : [rouge_1_kl_score],
    "bart" : [rouge_1_bart_score],
    "GPT" : [rouge_1_GPT_score]
})
bar(x=df.columns.values, y=df.values[0], title="Rouge-1")
```
